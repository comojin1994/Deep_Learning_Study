```python
import tensorflow as tf
import numpy as np
print(tf.__version__)
import os
from glob import glob
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_paths = glob(r'C:\Users\LG\Deep_Learning_Study\Tensorflow\dataset\mnist_png\training/*/*.png')
path = data_paths[0]
path
```

```python
data_paths = tf.io.matching_files(r'C:\Users\LG\Deep_Learning_Study\Tensorflow\dataset\mnist_png\training/*/*.png')
path = data_paths[0]
path
```

```python
gfile = tf.io.read_file(path)
image = tf.io.decode_image(gfile)
image.shape
```

```python
plt.imshow(image[:, :, 0], 'gray')
plt.show()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)
```

```python
inputs = image[tf.newaxis, ...]
inputs.shape
```

```python
result = next(iter(datagen.flow(inputs)))

plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.imshow(inputs[0, :, :, 0], 'gray')

plt.subplot(122)
plt.imshow(result[0, :, :, 0], 'gray')

plt.show()
```

```python
datagen = ImageDataGenerator(
    width_shift_range=0.3
)

outputs = next(iter(datagen.flow(inputs)))
plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.imshow(inputs[0, :, :, 0], 'gray')

plt.subplot(122)
plt.imshow(outputs[0, :, :, 0], 'gray')

plt.show()
```

```python
datagen = ImageDataGenerator(
    zoom_range=0.3
)

outputs = next(iter(datagen.flow(inputs)))
plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.imshow(inputs[0, :, :, 0], 'gray')

plt.subplot(122)
plt.imshow(outputs[0, :, :, 0], 'gray')

plt.show()
```

```python
datagen = ImageDataGenerator(
    preprocessing_function = lambda x: x * np.random.randn(1),
    rescale = 1./255.
)

outputs = next(iter(datagen.flow(inputs)))
plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.imshow(inputs[0, :, :, 0], 'gray')

plt.subplot(122)
plt.imshow(outputs[0, :, :, 0], 'gray')

plt.show()
```
